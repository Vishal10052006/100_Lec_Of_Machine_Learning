# Working on date

```python
date['date'] = pd.to_datetime(date['date'])    #converting object into datetime dataset
```

```python
date.info()
```

```python
#Extract year

date['date_year'] = date['date'].dt.year
date.head()
```

```python
date['date_month_no'] = date['date'].dt.month
date.head()
```

```python
date['date_month_name'] = date['date'].dt.month_name()
date.head()
```

```python
date['date_day'] = date['date'].dt.day
date.head()
```

```python
date['date_dayofweek'] = date['date'].dt.dayofweek
date.head()
```

```python
date['date_dateofname_name'] = date['date'].dt.day_name()
date.head()
```

```python
date['date_is_weekend'] = np.where(date['date_dateofname_name'].isin(['Sunday', 'Saturday']), 1,0)
date.drop(columns=['product_id', 'city_id', 'orders']).head()
```

# Ordinal Encoding

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('customer.csv')
df.sample(5)
```

```python
df = df.iloc[:,2:]
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.iloc[:,:2], df['purchased'], test_size = 0.2)
```

```python
# x_train
```

```python
# y_train
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories = [['Poor', 'Average', 'Good'],['School', 'UG', 'PG']])
```

```python
oe
```

```python
oe.fit(x_train)
```

```python
x_train = oe.transform(x_train)
x_test = oe.transform(x_test)
```

```python
x_train
```

```python
x_test
```

# Label Encoding

```python
df = pd.read_csv('customer.csv')
df.sample(5)
```

```python
# df.iloc[:,4]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.iloc[:,-1],
                                                     df['purchased'],
                                                     test_size = 0.2,
                                                     random_state = 0)
                   
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
le.fit(y_train)
```

```python
x_train = le.transform(x_train)
x_test = le.transform(x_test)
```
